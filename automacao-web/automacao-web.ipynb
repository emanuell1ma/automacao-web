# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/1mhXZ3JPAnekXP_4vX7Z_sJj35VWqayaR?usp=sharing
(https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema)

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time

# pyautogui.clik -> clicar o mouse
# pyautogui.write -> escrever um texto
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> aperta uma combinação de teclas

pyautogui.PAUSE = 1

# Passo 1: Acesso o sistema da empresa
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")
time.sleep(5)

# Passo 2: Fazer login no sistema
pyautogui.click(x=3266, y=426)
pyautogui.write("meu login")
pyautogui.click(x=3251, y=530)
pyautogui.write("minha senha")
pyautogui.click(x=3291, y=624)
time.sleep(5)

# Passo 3: Baixar a base de dados
pyautogui.click(x=2885, y=498, button="right")
pyautogui.click(x=3025, y=901)

time.sleep(5)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4: Calcular os indicadores
import pandas as pd

# Importar a base de dados
tabela = pd.read_csv(r"C:\Users\emanu\Downloads\Compras.csv", sep=";")
display(tabela)

# Calculo dos indicadores
# Total gasto -> somar a coluna valor final
total_gasto = tabela["ValorFinal"].sum()

# Quantidade -> somar a coluna quantidade
quantidade = tabela["Quantidade"].sum()

# Preço médio -> total gasto / quantidade
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyperclip
# Passo 5: Enviar e-mail para a diretoria/para o chefe

# Entrar no email
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/")
pyautogui.press("enter")

time.sleep(5)

# Clicar em escrever
pyautogui.click(x=2451, y=206)
pyautogui.click(x=2616, y=220)

# Preencher o email
pyautogui.write("emanuel.uberselect@gmail.com")
pyautogui.press("tab") # seleciona o email
pyautogui.press("tab") # passa pro campo de assunto
pyperclip.copy("Relatório de Compras")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") # passa pro campo do corpo de email

texto = f"""

Prezados,

Segue relatório  de compras,

Total Gasto: R$ {total_gasto:,.2f}
Quantidade de Produtos: {quantidade:,}
Preço Médio: R$ {preco_medio:,.2f}

Qualquer dúvida, estou a disposição

Att.,
Emanuel Lima
"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# Enviar
pyautogui.hotkey("ctrl", "enter")
```
